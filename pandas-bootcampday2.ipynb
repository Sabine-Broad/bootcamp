```python
df = pd.read_csv("california_housing.tsv", sep="\t", index_col=0)
```

```python
df
```

```python
california_ols = sm.OLS(df['MedHouseVal'], df.loc[:, ["MedInc"]]) #OLS is the ordinary least square
```

```python
california_ols_result = california_ols.fit()
```

```python
print(california_ols_result.summary())
```

```python
california_ols = sm.OLS(df['MedHouseVal'], sm.add_constant(df.loc[:, ["MedInc"]])) #OLS is the ordinary least square
california_ols_result = california_ols.fit()
print(california_ols_result.summary())
```

```python
california_ols = sm.OLS(df['MedHouseVal'], sm.add_constant(df.loc[:, ["MedInc"
                                                                      , "Latitude"]])) #OLS is the ordinary least square
california_ols_result = california_ols.fit()
print(california_ols_result.summary())
```
